```python
# Save time, use part of dataframe
df1 = df[['Date', 'Ticker', 'Return', 'SPY Return', 'Rf']]
df1.head(3)
```

# Rolling Window and Calculate Var & Cov

```python
df2 = Alpha_Beta.Rolling(df1, window=120)
```

```python
df3 = df2.dropna()
```

```python
df3
```

```python
df4 = df3.copy()[['Date', 'Ticker', 'Beta', 'Alpha']]
df4.head()
```

```python
df5 = pd.merge(df, df4, on = ['Date', 'Ticker'], how = 'outer')
```

```python
df5 
```

```python
df6 = df5.dropna()
```

```python
df6
```

```python
df6 = df6[(df6.Date >= '2018-03-01') & (df6.Date <= '2020-03-31')]
```

```python
Factors = ['1W', '1M', '3M', '6M']

# Combine any 2 factors
for i in range(4):
    for j in range(i+1, 4):
        new_factor = Factors[i] + " + " + Factors[j]
        Factors.append(new_factor)
        df6[new_factor] = (df6[Factors[i]] + df6[Factors[j]]) / 2 

# Combine any 3 factors
for i in range(4):
    for j in range(i+1, 4):
        for p in range(j+1, 4):
            new_factor = Factors[i] + " + " + Factors[j] + " + " + Factors[p]
            Factors.append(new_factor)
            df6[new_factor] = (df6[Factors[i]] + df6[Factors[j]] + df6[Factors[p]]) / 3 

Factors.append('Total Rank')
df6['Total Rank'] = (df6['1W'] + df6['1M'] + df6['3M'] + df6['6M']) / 4
```
