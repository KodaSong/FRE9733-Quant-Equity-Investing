# Rank Stocks in Each Factor

```python
# df = df[(df['Date'] >= '2018-03-01') & (df['Date'] <= '2020-03-31')]
# SPY = SPY[(SPY['Date'] >= '2018-03-01') & (SPY['Date'] <= '2020-03-31')]
# Rf = Rf[(Rf['Date'] >= '2018-03-01') & (Rf['Date'] <= '2020-03-31')]
```

```python
Dates = list(set(df['Date'].values))
Dates.sort()
Factors = ['PM_1W', 'PM_1M', 'PM_3M', 'PM_6M']

df_ = Rank(df, Factors[0], Dates)
for factor in Factors[1:]:
    df_ = Rank(df_, factor, Dates)
```

```python
df_.reset_index(inplace=True, drop=True)
df_ 
```

```python
df__ = df_.copy()
del df__['PM_1W']
del df__['PM_1M']
del df__['PM_3M']
del df__['PM_6M']

df__.rename(columns = {'PM_1W Rank':'1W', 'PM_1M Rank':'1M', 'PM_3M Rank':'3M', 'PM_6M Rank':'6M'}, inplace = True)
```

```python
del SPY['SPY_1W']
del SPY['SPY_1M']
del SPY['SPY_3M']
del SPY['SPY_6M']
```
