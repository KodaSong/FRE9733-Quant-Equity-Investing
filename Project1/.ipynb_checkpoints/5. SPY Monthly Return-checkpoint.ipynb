# Load Data

```python
df = pd.read_csv("D:\\NYU\Syllabus\Spring 2020\Quant Equity Invest\Project\Original Data\^GSPC.csv")
```

```python
df.head(3)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.set_index('Date', inplace=True)
df.head(3)
```

```python
Date = []
Date.append('1994-12')
for i in range(1995, 2016):
    for j in range(1, 13):
        date = str(i) + '-' + str(j)
        Date.append(date)
```

```python
Price = []
Return = []
N = len(Date)
for i in range(N):
    DF = df[Date[i]].copy()
    Price.append(DF.head(1)['Adj Close'].values[0])
```

```python
for i in range(len(Price)-1):
    Return.append(Price[i+1]/Price[i] - 1)
```

```python
Return 
```

```python
r = pd.Series(Return)
r 
```

```python
Date = Date[1:]
Date = pd.Series(Date)
```

```python
Date 
```

```python
df = pd.DataFrame(list(zip(Date, r)), columns=['Date', 'Return'])
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```
