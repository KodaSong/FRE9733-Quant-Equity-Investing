```python
Fundamentals = pd.read_csv('D:\\Python\Quant Equity Investing 9733\Minghua Code\Fundamentals_new.csv')
Monthly = pd.read_csv('D:\\Python\Quant Equity Investing 9733\Minghua Code\Monthly_new.csv')
```

```python
Monthly.set_index(['PERMNO','Date'], inplace=True)
Fundamentals.set_index(['PERMNO','Date'], inplace=True)
```

```python
df2 = pd.merge(Monthly, Fundamentals, on = ['PERMNO','Date'], how = 'outer')
```

```python
df2.drop(columns=['TICKER_y','COMNAM_y'], inplace=True)
df2.rename(columns={'TICKER_x':'TICKER', 'COMNAM_x':'COMNAM'}, inplace = True)
df2.sort_index(ascending=True, inplace=True)
df2 = df2.groupby(['PERMNO']).apply(lambda x: x.fillna(method = 'ffill')) ##fill missing values as if looking into the past. 
df2.dropna(inplace=True)
df2.reset_index(inplace=True)
df2.head(3)
```

```python
def delete_day(row):
    row = str(row)
    row = row[:7]
    return row

df2['Date'] = df2['Date'].apply(delete_day)
df2.head(3)
```

```python
print(df1.shape)
print(df2.shape)
```

```python
df1.set_index(['PERMNO', 'Date'], inplace=True)
df2.set_index(['PERMNO', 'Date'], inplace=True)
```

```python
df3 = pd.merge(df1, df2, on = ['PERMNO', 'Date'], how = 'outer')
df3.head()
```

```python
df3.dropna(subset=['TICKER_x'], inplace=True)
df3
```

```python
df3.dropna(subset=['Turnover'], inplace=True)
df3
```
