# Delete Day in Date

```python
def delete_day(row):
    row = row[:7]
    return row

Month['Date'] = Month['Date'].apply(delete_day)
Quarter['Date'] = Quarter['Date'].apply(delete_day)
```

# Set PERMNO as index

```python
Month.set_index(['PERMNO','Date'], inplace=True)
Quarter.set_index(['PERMNO','Date'], inplace=True)
```

```python
Month.head(3)
```

# Merge

```python
df = pd.merge(Month, Quarter, on = ['PERMNO', 'Date'], how = 'outer')
df
```

```python
indice = list(df[np.isnan(df['Price'])].index)
df = df.drop(indice)
```

# Delete Repeated Columns

```python
del df['TICKER_y']
del df['COMNAM_y']
df.rename(columns={'TICKER_x':'TICKER', 'COMNAM_x':'COMNAM'}, inplace = True)
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
def fill_values_in_same_quarter(Series, Series1):
    n = len(Series)
    Series = list(Series)  # Transform into list will be faster
    Series1 = list(Series1)
    for i in range(1,n):
        if np.isnan(Series[i]):
            if np.isnan(Series[i-1]) == False and Series1[i] == Series1[i-1]:
                Series[i] = Series[i-1]
            elif np.isnan(Series[i-2]) == False and Series1[i] == Series1[i-2]:
                Series[i] = Series[i-2]
    Series = pd.Series(Series)  # Go back to pd.Series so that can add into dataframe           
    return Series
```

```python
df.columns
```

```python
df['Current Assets'] = fill_values_in_same_quarter(df['Current Assets'], df['PERMNO'])
df['Current Liabilities'] = fill_values_in_same_quarter(df['Current Liabilities'], df['PERMNO'])
df['Assets'] = fill_values_in_same_quarter(df['Assets'], df['PERMNO'])
df['Operating Income'] = fill_values_in_same_quarter(df['Operating Income'], df['PERMNO'])
df['Liabilities'] = fill_values_in_same_quarter(df['Liabilities'], df['PERMNO'])
df['Revenue'] = fill_values_in_same_quarter(df['Revenue'], df['PERMNO'])
df['Equities'] = fill_values_in_same_quarter(df['Equities'], df['PERMNO'])
df['Leverage'] = fill_values_in_same_quarter(df['Leverage'], df['PERMNO'])
df['Current Ratio'] = fill_values_in_same_quarter(df['Current Ratio'], df['PERMNO'])
df['Shares Outstanding'] = fill_values_in_same_quarter(df['Shares Outstanding'], df['PERMNO'])
df['EPS_Basic'] = fill_values_in_same_quarter(df['EPS_Basic'], df['PERMNO'])
df['EPS_Operation'] = fill_values_in_same_quarter(df['EPS_Operation'], df['PERMNO'])
df['Net Income'] = fill_values_in_same_quarter(df['Net Income'], df['PERMNO'])
```

```python
df['Cash&Short Term'] = fill_values_in_same_quarter(df['Cash&Short Term'], df['PERMNO'])
df['Inventory'] = fill_values_in_same_quarter(df['Inventory'], df['PERMNO'])
df['Quick Ratio'] = fill_values_in_same_quarter(df['Quick Ratio'], df['PERMNO'])
df['Cash Ratio'] = fill_values_in_same_quarter(df['Cash Ratio'], df['PERMNO'])
```

```python
df['Growth_Assets'] = fill_values_in_same_quarter(df['Growth_Assets'], df['PERMNO'])
df['Growth_EPS_Basic'] = fill_values_in_same_quarter(df['Growth_EPS_Basic'], df['PERMNO'])
df['Growth_Liabilities'] = fill_values_in_same_quarter(df['Growth_Liabilities'], df['PERMNO'])
df['Growth_Operating_Income'] = fill_values_in_same_quarter(df['Growth_Operating_Income'], df['PERMNO'])
df['Growth_EPS_Operation'] = fill_values_in_same_quarter(df['Growth_EPS_Operation'], df['PERMNO'])
df['Growth_Net Income'] = fill_values_in_same_quarter(df['Growth_Net Income'], df['PERMNO'])
df['Growth_Revenue'] = fill_values_in_same_quarter(df['Growth_Revenue'], df['PERMNO'])
df['Growth_Equities'] = fill_values_in_same_quarter(df['Growth_Equities'], df['PERMNO'])
df['Growth_Leverage'] = fill_values_in_same_quarter(df['Growth_Leverage'], df['PERMNO'])

df
```
