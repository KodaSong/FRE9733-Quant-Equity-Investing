# Import Packages

```python
import numpy as np 
import pandas as pd 
```

# Load Data

```python
df_lag = pd.read_csv("D:\\NYU\Syllabus\Spring 2020\Quant Equity Invest\Project\Original Data\Monthly_lag.csv")
df_return = pd.read_csv("D:\\NYU\Syllabus\Spring 2020\Quant Equity Invest\Project\Original Data\Monthly_Return.csv")
```

# Only reserve Year-Month

```python
def delete_day(row):
    row = row[:7]
    return row

df_lag['Date'] = df_lag['Date'].apply(delete_day)
df_return['Date'] = df_return['Date'].apply(delete_day)
```

# Merge

```python
df_lag.set_index(['PERMNO','Date'], inplace=True)
df_return.set_index(['PERMNO','Date'], inplace=True)
```

```python
df = pd.merge(df_lag, df_return, on = ['PERMNO', 'Date'], how = 'outer')
df
```

# Drop missing rows

```python
indice = list(df[np.isnan(df['Price'])].index)
df = df.drop(indice)
```
