# Rank for factor

```python
# Transform Date into datetime type
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

SPY['Date'] = pd.to_datetime(SPY['Date'])
SPY.set_index('Date', inplace=True)

# Construct Date List
Date = []
for i in range(1995, 2016):
    for j in range(1, 13):
        date = str(i) + '-' + str(j)
        Date.append(date)
```

```python
def Rank(df_, factor, factor_rank, Date):
    df = df_.copy()
        
    # First, we select the first date's dataframe and rank
    dataframe = df[Date[0]].copy()
    dataframe = dataframe.sort_values(by = [factor], axis = 0, ascending = False)  # Sort by Turnover
    dataframe[factor_rank] = range(len(dataframe))  
    dataframe[factor_rank] += 1 
    
    for i in range(1, len(Date)):
        DF = df[Date[i]].copy()  # Divide the big df into small DF by date
        DF = DF.sort_values(by = [factor], axis = 0, ascending = False)  # Sort by Turnover
        DF[factor_rank] = range(len(DF))  
        DF[factor_rank] += 1 
        dataframe = pd.concat([dataframe, DF])
    
    return dataframe
```

```python
df = Rank(df, 'Turnover', 'Turnover_rank', Date)
df = Rank(df, 'Dollar Turnover', 'Dollar_rank', Date)
df = Rank(df, 'Current Ratio', 'Current_rank', Date)
df = Rank(df, 'Cash Ratio', 'Cash_rank', Date)
df = Rank(df, 'Quick Ratio', 'Quick_rank', Date)
```

# Select Stocks based on Rank

```python
class Solve():
    def __init__(self, df, SPY, Date):
        self.df = df
        self.SPY = SPY
        self.Date = Date
        
    def Get_Return(self, Factor):
        self.df_ = self.df.copy()
        self.df_['Rank'] = self.df_[Factor]
        Return = []  # Used to store excess return for each month
        Excess = []  # Used to store excess return for each month
        for date in Date:
            DF = self.df_[date].copy()  # Divide the big df into small DF by date
            N = int(len(DF)/10)   # We decide to long top 10% of universe and short bottom 10%
            DF = DF.sort_values(by = ['Rank'], axis = 0, ascending = True)  # Sort by Rank
            #profit = ( DF.iloc[:N]['Return'].mean() - DF.iloc[-N:]['Return'].mean() ) / 2 
            profit = DF.iloc[int(0.1*N):int(1.1*N)]['Return'].mean()
            #profit = DF.iloc[-N:]['Return'].mean()
            profit_SPY = self.SPY[date]['Return'].values[0]

            Return.append( profit )
            Excess.append( profit - profit_SPY )

        self.Result = pd.DataFrame(list(zip(Date, Return, Excess)), columns=['Date', 'Return', 'Excess']) 
        self.Result['Date'] = pd.to_datetime(self.Result['Date'])
#         self.Return = pd.DataFrame(list(zip(Date, Return)), columns=['Date', 'Return']) 
#         self.Excess = pd.DataFrame(list(zip(Date, Excess)), columns=['Date', 'Excess'])
#         self.Return['Date'] = pd.to_datetime(self.Return['Date'])
#         self.Excess['Date'] = pd.to_datetime(self.Excess['Date'])

    def Return_Summary(self):
        R = self.Result['Return']
        S = self.SPY['Return']
        E = self.Result['Excess']        
        print("Summary for Return:"); print(R.describe(), "\n")
        print("Summary for SPY Return:"); print(S.describe(), "\n")
        print("Summary for Excess Return:"); print(E.describe(), "\n")

    def Plot_hist_Excess(self):
        plt.figure(figsize=(16,8))
        
        ax1 = plt.subplot(3,3,1)
        plt.hist(self.Result['Return'], bins=20)
        plt.ylabel('Frequency')
        plt.title('Distribution of Return')
        
        ax2 = plt.subplot(3,3,2)
        plt.hist(self.SPY['Return'], bins=20)
        plt.ylabel('Frequency')
        plt.title('Distribution of SPY Return')
        
        ax3 = plt.subplot(3,3,3)
        plt.hist(self.Result['Excess'], bins=20)
        plt.ylabel('Frequency')
        plt.title('Distribution of Excess Return')
        plt.show()
        
    def T_Test(self):
        from scipy import stats
        print(stats.ttest_1samp(self.Result['Excess'], 0), "\n")
        
    def Plot_Cum_Return(self):
        Return = self.Result['Return'].copy(); Return +=1; Return_ = Return.cumprod() - 1
        Excess = self.Result['Excess'].copy(); Excess +=1; Excess_ = Excess.cumprod() - 1
        Rm = self.SPY.copy()['Return'].values; Rm +=1; Rm_ = Rm.cumprod() - 1
    
        self.dfxxx = pd.DataFrame(list(zip(self.Date, Return_, Rm_, Excess_)), columns=['Date', 'Return', 'SPY', 'Excess'])
        self.dfxxx['Date'] = pd.to_datetime(self.dfxxx['Date'])
        self.dfxxx.set_index('Date', inplace=True)
        print(self.dfxxx)
        
        plt.figure(figsize=(12,6))
        plt.plot(self.dfxxx['Return'], label = 'Return')
        plt.plot(self.dfxxx['SPY'], label = 'SPY Return')
        plt.plot(self.dfxxx['Excess'], label = 'Excess Return')
        plt.legend()
        plt.show()
```

```python
Object = Solve(df, SPY, Date)
```

```python
Object.Get_Return('Turnover_rank')
Object.Return_Summary()
Object.Plot_hist_Excess()
Object.T_Test()
Object.Plot_Cum_Return()
```
