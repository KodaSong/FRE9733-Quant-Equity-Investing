# Do some Calculations

## First, Month Dataset

```python
Month = Month[Month['Price'] > 0]
Month['Turnover'] = Month['Volume'] / Month['Shares Outstanding']
Month['Dollar Turnover'] = Month['Turnover'] * Month['Price']
Month.head(3)
```

## Second, Quarter Dataset

```python
Quarter['Equities'] = Quarter['Assets'] - Quarter['Liabilities']
Quarter['Leverage'] = Quarter['Liabilities'] / Quarter['Assets']
Quarter['Current Ratio'] = Quarter['Current Assets'] / Quarter['Current Liabilities']
Quarter['Quick Ratio'] = (Quarter['Current Assets'] - Quarter['Inventory']) / Quarter['Current Liabilities']
Quarter['Cash Ratio'] = Quarter['Cash&Short Term'] / Quarter['Current Liabilities']

Quarter.head(3)
```
